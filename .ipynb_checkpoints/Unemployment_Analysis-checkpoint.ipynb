```python
avg_unemployment = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
```

## Region-wise Average Unemployment Rate

This chart shows the average unemployment rate across different regions of India.

```python
avg_unemployment.plot(kind='bar', figsize=(12,6))

plt.title("Average Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Average Unemployment Rate (%)")
plt.xticks(rotation=90)

plt.show()
```

## Observation

- The bar chart shows the average unemployment rate for each region.
- Some regions have a higher average unemployment rate than others.
- This helps us compare unemployment levels across different regions.

```python
states = ['Maharashtra', 'Delhi', 'Karnataka']

filtered_df = df[df['Region'].isin(states)]

filtered_df.head()
```

## Time Series Analysis

This line chart shows the unemployment rate over time for Maharashtra, Delhi, and Karnataka. It helps visualize trends and highlights the impact of COVID-19 on unemployment.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

for state in states:
    state_data = filtered_df[filtered_df['Region'] == state]
    plt.plot(
        state_data['Date'],
        state_data['Estimated Unemployment Rate (%)'],
        marker='o',
        label=state
    )

plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.legend()
plt.grid(True)

plt.show()
```

## Observation

- The line chart compares unemployment trends over time for Maharashtra, Delhi, and Karnataka.
- The unemployment rate varies across states and changes noticeably over different months.
- This visualization helps identify trends and compare regional unemployment patterns.

```python
top10 = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top10.plot(kind='bar')

plt.title("Top 10 States with Highest Average Unemployment Rate")
plt.xlabel("State")
plt.ylabel("Average Unemployment Rate (%)")
plt.xticks(rotation=45)

plt.show()
```

## Observation

- Tripura has the highest average unemployment rate among the selected states.
- Haryana and Jharkhand also show relatively high unemployment.
- The chart highlights the states most affected by unemployment during the observed period.

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation = df[['Estimated Unemployment Rate (%)',
                  'Estimated Employed',
                  'Estimated Labour Participation Rate (%)']].corr()

plt.figure(figsize=(8,6))

sns.heatmap(correlation,
            annot=True,
            cmap='coolwarm',
            fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation = df[['Estimated Unemployment Rate (%)',
                  'Estimated Employed',
                  'Estimated Labour Participation Rate (%)']].corr()

plt.figure(figsize=(8,6))

sns.heatmap(correlation,
            annot=True,
            cmap='coolwarm',
            fmt=".2f")

plt.title("Correlation Heatmap")
plt.show()
```

### Observation

- The heatmap shows the correlation between unemployment rate, estimated employed, and labour participation rate.
- A positive value indicates that two variables increase together.
- A negative value indicates that one variable decreases as the other increases.
- This visualization helps understand the relationship between employment-related indicators.

```python
pre_covid = df[df['Date'] < '2020-03-01']
post_covid = df[df['Date'] >= '2020-03-01']
```

```python
pre_avg = pre_covid['Estimated Unemployment Rate (%)'].mean()
post_avg = post_covid['Estimated Unemployment Rate (%)'].mean()

print("Average Unemployment Rate Before COVID :", round(pre_avg, 2))
print("Average Unemployment Rate After COVID  :", round(post_avg, 2))
```

```python
import matplotlib.pyplot as plt

periods = ['Pre-COVID', 'Post-COVID']
rates = [pre_avg, post_avg]

plt.figure(figsize=(6,5))
plt.bar(periods, rates)

plt.title("Pre-COVID vs Post-COVID Average Unemployment Rate")
plt.xlabel("Period")
plt.ylabel("Average Unemployment Rate (%)")

plt.show()
```

### Observation

- The average unemployment rate increased after the COVID-19 outbreak.
- The lockdown significantly affected employment across different regions of India.
- The comparison clearly shows the economic impact of the pandemic on unemployment.

## Pre-COVID vs Post-COVID Comparison

```python
pre_covid = df[df['Date'] < '2020-03-01']
post_covid = df[df['Date'] >= '2020-03-01']

print("Average Unemployment Rate Before COVID:")
print(pre_covid['Estimated Unemployment Rate (%)'].mean())

print("\nAverage Unemployment Rate After COVID:")
print(post_covid['Estimated Unemployment Rate (%)'].mean())
```

### Observation

- The average unemployment rate increased after the COVID-19 outbreak.
- This indicates the significant impact of the pandemic on employment in India.
